```python
import pandas as pd
import numpy as np
```

```python
movies = pd.read_csv(r'C:\Users\jaiad.APPAUMMI\Documents\MOVIE RECOMMENDATION SYSTEM\Dataset\tmdb_5000_movies.csv')
credits = pd.read_csv(r'C:\Users\jaiad.APPAUMMI\Documents\MOVIE RECOMMENDATION SYSTEM\Dataset\tmdb_5000_credits.csv')
```

```python
movies = movies.merge(credits, on = 'title')
```

```python
movies.head()
```

```python
movies.shape
```

```python
movies = movies[['movie_id','title','keywords','genres','overview','cast','crew']]
```

```python
movies.head()
```

```python
movies.shape
```

```python
import ast

def change(text):
    l=[]
    for i in ast.literal_eval(text):
        l.append(i['name'])
        
    return l
```

```python
movies['genres'] = movies['genres'].apply(change)
```

```python
movies.head()
```

```python
movies.shape
```

```python
import ast

def context(text):
    l=[]
    for i in ast.literal_eval(text):
        l.append(i['name'])
        
    return l
```

```python
movies['keywords'] = movies['keywords'].apply(context)
```

```python
movies.head()
```

```python
movies.shape
```

```python
import ast

def value(text):
    l=[]
    counter = 0
    for i in ast.literal_eval(text):
        if counter<3:
            l.append(i['name'])
        counter+=1
        
    return l
```

```python
movies['cast'] = movies['cast'].apply(value)
```

```python
movies.head()
```

```python
import ast

def kind(text):
    l=[]
    for i in ast.literal_eval(text):
        l.append(i['name'])
        
    return l
```

```python
movies['crew'] = movies['crew'].apply(kind)
```

```python
movies.head()
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.isnull().sum()
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
def remove(word):
    l=[]
    for i in word:
        l.append(i.replace(" ", ""))
                
    return l
```

```python
movies['cast'] = movies['cast'].apply(remove)
movies['keywords'] = movies['keywords'].apply(remove)
movies['genres'] = movies['genres'].apply(remove)
movies['crew'] = movies['crew'].apply(remove)
```

```python
movies.head()
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies.head()
```

```python
movies.iloc[0]['tags']
```

```python
newdf = movies[['movie_id','title','tags']]
```

```python
newdf.head()
```

```python
newdf['tags'] = newdf['tags'].apply(lambda x: " ".join(x))
```

```python
newdf.head()
```

```python
import nltk
from nltk import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
def stems(text):
    l=[]
    for i in text.split():
        l.append(ps.stem(i))
        
        
    return " ".join(l)
```

```python
newdf['tags'] = newdf['tags'].apply(stems)
```

```python
newdf.iloc[0]['tags']
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
vector = cv.fit_transform(newdf['tags']).toarray()
```

```python
vector
```

```python
vector.shape
```

```python
cv.vocabulary_ #gives the no of words in each sentence
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
sim = cosine_similarity(vector)
```

```python
sim
```

```python
sim.shape
```

```python
newdf[newdf['title'] == 'Spider-Man'].index[0]
```

```python
newdf.head(160) #reference
```

```python
def recommend(movie):
    index = newdf[newdf['title'] == movie].index[0]
    distances = sorted(list(enumerate(sim[index])), reverse= True, key = lambda x:x[1])
    for i in distances[1:10]:
        print(newdf.iloc[i[0]].title)
```

```python
recommend('Avatar') #in dataframe
```

```python
recommend('Harry Potter') #not in dataframe
```
